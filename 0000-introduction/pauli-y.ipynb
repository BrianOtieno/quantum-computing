<center><h1>Pauli-Y</h1></center>

### Mathematical Representation
The **Pauli-Y gate** acts on a single qubit and has the following matrix form:
$$
Y = \begin{bmatrix} 
0 & -i \\ 
i & 0 
\end{bmatrix}
$$

It transforms the basis states $|0\rangle$ and $|1\rangle$ as follows:
$$
Y|0\rangle = i|1\rangle, \quad Y|1\rangle = -i|0\rangle
$$
Note the introduction of imaginary coefficients, which distinguishes the Pauli-Y gate from Pauli-X by adding a phase factor.

---

### Quantum State Transformation
For a general qubit state:
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$$

Applying the Pauli-Y gate gives:
$$
Y|\psi\rangle = -i\beta|0\rangle + i\alpha|1\rangle
$$

This shows that the Pauli-Y gate swaps the amplitudes of $|0\rangle$ and $|1\rangle$ while multiplying them by $i$ and $-i$, respectively, introducing a 90-degree phase shift.

---

### Properties of Pauli-Y
1. **Hermitian:** $Y^\dagger = Y$  
   (The conjugate transpose equals itself, as $Y$ is its own Hermitian conjugate.)
2. **Unitary:** $Y^\dagger Y = I$  
   (The product of $Y$ with its conjugate transpose is the identity matrix, ensuring reversibility.)
3. **Involutory:** $Y^2 = I$  
   (Applying $Y$ twice returns the original state, since $Y$ is its own inverse.)

These properties align with those of Pauli-X, making Pauli-Y another reversible operation crucial for quantum computing.

### Qiskit Implementation
Here’s how to implement and visualize the Pauli-Y gate using Qiskit (see the code cells below for the implementation steps):

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector
```

```python
# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
```

```python
# Apply Pauli-Y gate to qubit 0
qc.y(0)
```

```python
# Visualize the circuit
print("Quantum Circuit:")
print(qc.draw('text'))
```

```python
# Save the statevector to retrieve it after simulation
qc.save_statevector()

# Simulate the effect of Pauli-Y gate on initial |0⟩ state
simulator = AerSimulator()
result = simulator.run(qc).result()
statevector = result.get_statevector()
```

```python
# Print the resulting state vector
print("\nStatevector after Pauli-Y gate:")
print(statevector)
```

```python
# Bloch sphere visualization
plot_bloch_multivector(statevector)
```

---

### Explanation of Results
- **Statevector:** The transformation $Y|0\rangle = i|1\rangle$ results in the statevector `[0.+0.j, 0.+1.j]`. The imaginary unit $i$ reflects the phase introduced by Pauli-Y.
- **Bloch Sphere:** The Pauli-Y gate rotates the state by $\pi$ radians (180 degrees) around the $y$-axis, combined with a phase shift. From $|0\rangle$ (north pole, along $+z$), it moves to a state proportional to $|1\rangle$ with a phase of $i$, visualized as a point on the positive $y$-axis in the equatorial plane.
- **Comparison to Pauli-X:** Unlike Pauli-X, which flips $|0\rangle$ to $|1\rangle$ without phase changes, Pauli-Y introduces a relative phase of $i$ and $-i$, making it a rotation with a twist in the complex plane.
