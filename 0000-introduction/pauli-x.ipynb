<center><h1>Pauli-X</h1></center>

### **Mathematical Representation**
The **Pauli-X gate** acts on a single qubit and has the following matrix form:
$$
X = \begin{bmatrix} 
0 & 1 \\ 
1 & 0 
\end{bmatrix}
$$

It swaps the amplitudes of the $|0\rangle$ and $|1\rangle$ states:

$$
X|0\rangle = |1\rangle, \quad X|1\rangle = |0\rangle
$$

### **Quantum State Transformation**

For a general qubit state:
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle,
$$

applying the Pauli-X gate gives:
$$
X|\psi\rangle = \alpha|1\rangle + \beta|0\rangle
$$

### **Properties of Pauli-X**
1. **Hermitian:** $X^\dagger = X$
2. **Unitary:** $X^\dagger X = I$
3. **Involutory:** $X^2 = I$

These properties make the Pauli-X gate a reversible operation, critical for quantum computing.

### **Qiskit Implementation**
Here’s how to implement and visualize the Pauli-X gate using Qiskit:



```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector
```

```python
# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
```

```python
# Apply Pauli-X gate
qc.x(0)
```

```python
# Visualize the circuit
print("Quantum Circuit:")
print(qc.draw('text'))
```

```python
#Visualize using mpl
qc.draw('mpl')
```

```python
# Save the statevector to retrieve it after simulation
qc.save_statevector()
```

```python
# Simulate the effect of Pauli-X gate on initial |0⟩ state
simulator = AerSimulator()
result = simulator.run(qc).result()
statevector = result.get_statevector()
```

```python
# Retrieve the saved statevector
statevector = result.get_statevector()
```

```python
# Print the resulting state vector
print("\nStatevector after Pauli-X gate:")
print(statevector)
```

```python
# Bloch sphere visualization
plot_bloch_multivector(statevector)
```
